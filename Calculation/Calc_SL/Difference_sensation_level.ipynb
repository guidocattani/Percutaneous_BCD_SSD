# Calculation of difference BC - AC sensation levels

```python
from pathlib import Path
import pandas as pd
```

```python
def files_to_dict(l):
    d = dict()
    for p in l:
        df = pd.read_csv(p)
        df.set_index('Study_ID', inplace=True)
        sp = str(p)
        ls = sp.split('/')
        fn = ls[-1]
        lf = fn.split('.')
        r = lf[0]
        print(r)
        element = {r : df}
        d.update(element)
    #sorted_dict = {k: d[k] for k in sorted(d)}
    return d
```

```python
def list_of_files(p = '/media/guido/LACIE/Cingle_Guido/Master/Implant/Calculated_data'):
    lbc = list(Path(p).glob('BC_SL_*.csv'))    
    lac = list(Path(p).glob('AC_SL_*.csv'))

    dbc = files_to_dict(lbc)
    dac = files_to_dict(lac)
    
    t = (dbc, dac)
    return t 
```

```python
list_of_files()
```

```python
def calc_diff_sl(l):
    dbc, dac = list_of_files()
    dres = dict()
    for t in l:
        bc = dbc[t[0]]
        ac = dac[t[1]]
        dsl = bc.sub(ac, axis = 1)
        dsl = dsl.round(1)
        dsl = ((dsl*10).astype(int).astype(float))/10
        list_of_strings = t[0].split('_')
        if list_of_strings[3] == '0deg': pos = 'S_front'
        if list_of_strings[3] == '90deg': pos = 'S_BCDside'
        if list_of_strings[3] == '270deg': pos = 'S_bestearside'
        name = 'Diff_SL_' + list_of_strings[2] + '_' + pos
        dres.update({name : dsl})
    return dres
```

```python
def diff_sl_to_dict():
    # return dict with BC sensation levels for 0, 90 and 270 degrees
    
    dres = dict()
    
    l65 = [('BC_SL_65dB_0deg', 'AC_SL_65dB_0deg'), ('BC_SL_65dB_90deg', 'AC_SL_65dB_270deg'), ('BC_SL_65dB_270deg', 'AC_SL_65dB_90deg')]  # combinations of degrees for 65 dB
    d65 = calc_diff_sl(l65)
    
    l55 = [('BC_SL_55dB_0deg', 'AC_SL_55dB_0deg'), ('BC_SL_55dB_90deg', 'AC_SL_55dB_270deg'), ('BC_SL_55dB_270deg', 'AC_SL_55dB_90deg')]  # combinations of degrees for 55 dB
    d55 = calc_diff_sl(l55)
    
    dres = {**d65, **d55}
    
    return dres
```

```python
ddsl = diff_sl_to_dict()
ddsl['Diff_SL_65dB_S_front']
```

```python
def save_results(pout = '/media/guido/LACIE/Cingle_Guido/Master/Implant/Calculated_data/'):
    print('INPUT FILES')
    d = diff_sl_to_dict()
    print()
    print('OUTPUT FILES')
    for k in d:
        fout = pout + k +'.csv'
        print(fout)
        d[k].to_csv(fout)
```

```python
if __name__ == '__main__':
    save_results()
```
