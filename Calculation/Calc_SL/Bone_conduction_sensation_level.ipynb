```python
from pathlib import Path
import pandas as pd
```

```python
def band_names():
    # return list with column names for results table
    frequencies = [125, 160, 200, 250, 315, 400, 500, 630, 800, 1000, 1250, 1600, 2000, 2500, 3150, 4000, 5000, 6300, 8000]
    l = list()
    for f in frequencies:
        st = 'f_' + str(f) + '_Hz'
        l.append(st)
    return l
```

```python
def change_f_columns_names(df):
    df = df.T
    df.insert(0, 'band_name', band_names())
    df.set_index('band_name', drop=True, inplace=True)
    df.index.name = None
    df = df.T
    return df
```

```python
def read_BC_thresholds():
    # function to read threshold in dB FL
    fin = '/media/guido/LACIE/Cingle_Guido/Master/Implant/Calculated_data/BC_dir_thr_third.csv'
    pin = Path(fin)   
    df = pd.read_csv(pin)
    df.set_index('Study_ID', drop=True, inplace=True)
    df = df.fillna(pd.NA)
    return df
```

```python
read_BC_thresholds()
```

```python
def read_HRTF_BAHA():
    # function to read HRTF data for the BC path, collected by Stenfelt
    f_in = '/media/guido/LACIE/Cingle_Guido/Master/Implant/Primary_data/constants.xlsx'
    p_in = Path(f_in)   
    col_to_use = list(range(20))
    df = pd.read_excel(p_in, sheet_name='HRTF_BAHA_Stenfelt', header=0, nrows=3, usecols=col_to_use)
    df = df.rename(columns={'Unnamed: 0' : 'Angle'})
    df = df.set_index(['Angle'])
    df = df.round(1)
    df = change_f_columns_names(df)
    s0 = pd.Series(df.iloc[0])
    s90 = pd.Series(df.iloc[1])
    s270 = pd.Series(df.iloc[2])
    return (s0, s90, s270)
```

```python
def read_BCD_output_65():
    # function to read 
    f_in = '/media/guido/LACIE/Cingle_Guido/Master/Implant/Primary_data/output_BCD_65dB.csv'
    p_in = Path(f_in)   
    df = pd.read_csv(p_in)
    df.set_index('Study_ID', drop=True, inplace=True)
    return df
```

```python
read_BCD_output_65()
```

```python
def read_BCD_output_55():
    # function to read threshold in dB FL
    f_in = '/media/guido/LACIE/Cingle_Guido/Master/Implant/Primary_data/output_BCD_55dB.csv'
    p_in = Path(f_in)   
    df = pd.read_csv(p_in)
    df.set_index('Study_ID', drop=True, inplace=True)
    return df
```

```python
read_BCD_output_55()
```

```python
def sl_65():
    # return list with BC sensation levels for 0, 90 and 270 degrees
    bcd_out = read_BCD_output_65()
    bc_thr = read_BC_thresholds()
    htrf = read_HRTF_BAHA()
    bn = band_names()
    sl1 = bcd_out.sub(bc_thr, axis=1)
    
    l = list()
    
    for i in range(len(htrf)):        
        sl = sl1.add(htrf[i], axis=1)
        sl = sl.round(decimals=1)
        for f in ['f_125_Hz', 'f_160_Hz', 'f_200_Hz']: sl[f] = sl['f_250_Hz']   
        sl = ((sl*10).astype(int).astype(float))/10
        sl = sl[bn]
        l.append(sl)
    return l
```

```python
sl65 = sl_65()
sl55 = sl_55()
```
